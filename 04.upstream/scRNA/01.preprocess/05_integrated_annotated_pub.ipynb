```r
# Load annotated integrated object 
obj <- readRDS('/mnt/18T/chibao/gliomas/data_official/01_integrated_obj/new/obj/02_rpca_integrated_orig.ident_annotated.rds')
obj
obj@meta.data |> head(2)
```

```r
# Set suitable idents
Idents(obj) <- obj@meta.data$general_cell_type
```

```r
# Find markers for each cell type cluster
all_markers <- FindAllMarkers(obj, 
                              assay = 'SCT',
                              only.pos = TRUE,
                              min.pct = 0.25,
                              logfc.threshold = 0.25,
                              test.use = 'wilcox')
all_markers |> head(2)
```

```r
# Save markers for each cell type cluster
all_markers <- all_markers |> arrange(desc(avg_log2FC))
write.csv(all_markers, '/mnt/18T/chibao/gliomas/data_official/01_integrated_obj/new/markers/cell_type_markers.csv')
```
